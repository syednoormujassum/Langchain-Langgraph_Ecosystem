## Data Ingestion Pipeline LangChain

```python
## Text Loader
from langchain_community.document_loaders import TextLoader
import os
from pathlib import Path

loader=TextLoader((Path.cwd().parent)/"sample.txt")
doc_text=loader.load()
doc_text
```

```python
## reading a pdf file
from langchain_community.document_loaders import PyPDFLoader
pdf_doc=PyPDFLoader((Path.cwd().parent)/"attention_is_all_you_need.pdf")
pdf_text=pdf_doc.load()
pdf_text
```

```python
## Web Based Loader
from langchain_community.document_loaders import WebBaseLoader
import bs4
web_text_loader=WebBaseLoader(("https://en.wikipedia.org/wiki/Anthropic/"),
                bs_kwargs=dict(parse_only=bs4.SoupStrainer(class_=("post-title","post-content","post-header"))))
web_text_loader.load()
```

```python
## Arxiv
from langchain_community.document_loaders import ArxivLoader

# Supports all arguments of `ArxivAPIWrapper`
loader = ArxivLoader(
    query="1706.03762",
    load_max_docs=2,
    # doc_content_chars_max=1000,
    # load_all_available_meta=False,
    # ...
)
```

```python
arxiv_docs = loader.load()
arxiv_docs
```

```python
## wikipedia

from langchain_community.document_loaders import WikipediaLoader
wiki_docs = WikipediaLoader(query="Anthropic", load_max_docs=2)
wiki_docs.load()
```

```python
##SQL
from langchain_community.document_loaders import SQLDatabaseLoader
from langchain_community.utilities import SQLDatabase

db = SQLDatabase.from_uri(
    "mysql+pymysql://root:root1234@localhost/dev"
)
```

```python
sql_loader=SQLDatabaseLoader(
    db=db,
    query="select * from users"
)

sql_docs=sql_loader.load()
```

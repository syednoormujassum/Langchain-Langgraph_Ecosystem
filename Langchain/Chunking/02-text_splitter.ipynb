```python
import os
from pathlib import Path
from langchain_community.document_loaders import PyPDFLoader
#print(Path.cwd().parent)
pdf_doc=PyPDFLoader((Path.cwd().parent)/"attention_is_all_you_need.pdf")
pdf_text=pdf_doc.load()
pdf_text
```

```python
### Recursively Text Splitting by characters



text_splitter=RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=100)
final_pdf_doc=text_splitter.split_documents(pdf_text)
final_pdf_doc
```

# Embedding 
##### Process of converting Text chunks into vectors 

```python
import os
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from pathlib import Path
from dotenv import load_dotenv
load_dotenv()
```

```python
os.environ["OPENAI_API_KEY"]=os.getenv("OPENAI_API_KEY")
```

```python
from langchain_openai import OpenAIEmbeddings
embedding=OpenAIEmbeddings(model="text-embedding-3-large")
embedding
```

```python
text="This is a tutorial of OPENAI Embedding"
embedding.embed_query(text)
```

```python
len(embedding.embed_query(text))
```

```python
embedding_1024=OpenAIEmbeddings(model="text-embedding-3-large",dimensions=1024)
```

```python
embedding_1024.embed_query(text)
```

```python
len(embedding_1024.embed_query(text))
```

```python
loader=TextLoader((Path.cwd().parent)/"sample.txt")
doc_text=loader.load()
doc_text
```

```python
text_chunk=RecursiveCharacterTextSplitter(chunk_size=100,chunk_overlap=30)
final_text_chunks=text_chunk.split_documents(doc_text)
final_text_chunks
```

```python
#    embedding=OpenAIEmbeddings(model="text-embedding-3-large")
#    embedding.embed_documents([final_text_chunks])
```

```python
# Vector Embedding and Vector Store DB 

from langchain_community.vectorstores import Chroma

db=Chroma.from_documents(final_text_chunks,embedding_1024)
db
```

```python
query="Toronto."

retreiveed_results=db.similarity_search(query)
```
